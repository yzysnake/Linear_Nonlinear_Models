```python
from sympy import symbols, diff, exp
import numpy as np
from scipy.optimize import newton
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the symbol
x = symbols('x')

# Define the function f(x) = 3x^2 - 2x - 4
f1 = 3*x**2 - 2*x - 4

# Calculate its derivative f'(x)
f1_prime = diff(f1, x)
f1, f1_prime
```

```python
# Define the function and its derivative as lambda functions for numerical computation
f1_num = lambda x: 3*x**2 - 2*x - 4
f1_prime_num = lambda x: 6*x - 2

# Choose an initial guess
initial_guess = 1.5

# Define the Newton-Raphson method function
def newton_raphson(func, derivative, initial, tol=1.0e-6, max_iter=100):
    x_n = initial
    iter_data = []  # to store iteration data
    
    for n in range(max_iter):
        f_n = func(x_n)
        f_prime_n = derivative(x_n)
        if f_prime_n == 0:
            raise ZeroDivisionError(f"Derivative zero for x = {x_n}")
        x_next = x_n - f_n / f_prime_n
        err = abs(x_next - x_n)
        
        iter_data.append((x_n, f_n, f_prime_n, err))
        
        if err < tol:
            break
        x_n = x_next
        
    return x_n, iter_data

# Apply the Newton-Raphson method to our function
root, iter_data = newton_raphson(f1_num, f1_prime_num, initial_guess)

# Create a DataFrame to store iteration data
iter_df = pd.DataFrame(iter_data, columns=['x', 'f(x)', 'f\'(x)', 'err'])

# Plot f(x) around the solution
x_values = np.linspace(root - 2, root + 2, 400)
y_values = f1_num(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = 3x^2 - 2x - 4')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Root ≈ {root:.4f}')
plt.title('Plot of f(x) around the root')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

root, iter_df.head()  # Show the root and the first few iterations data
```

```python
# Define the function and its derivative as lambda functions for numerical computation
f2_num = lambda x: np.exp(x) - x**4
f2_prime_num = lambda x: np.exp(x) - 4*x**3

# Choose an initial guess. For this function, we'll pick a value close to where the function crosses the x-axis
initial_guess = 0

# Apply the Newton-Raphson method to our second function
root_2, iter_data_2 = newton_raphson(f2_num, f2_prime_num, initial_guess)

# Create a DataFrame to store iteration data for the second function
iter_df_2 = pd.DataFrame(iter_data_2, columns=['x', 'f(x)', 'f\'(x)', 'err'])

# Plot f(x) around the solution for the second function
x_values_2 = np.linspace(root_2 - 1, root_2 + 1, 400)
y_values_2 = f2_num(x_values_2)

plt.figure(figsize=(10, 6))
plt.plot(x_values_2, y_values_2, label='f(x) = e^x - x^4')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(root_2, color='red', linestyle='--', label=f'Root ≈ {root_2:.4f}')
plt.title('Plot of f(x) around the root')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

root_2, iter_df_2.head()  # Show the root and the first few iterations data
```
