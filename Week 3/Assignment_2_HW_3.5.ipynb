```python
import numpy as np
import matplotlib.pyplot as plt
from math import log, pi, comb, factorial, log10
from scipy.stats import norm, binom, poisson
```

```python
# 1

# Initialize variables
mu = 0.1
std = 0.15
y_predictor = np.arange(-0.5, 0.71, 0.01)

# Calculations
T = y_predictor / std
eta = mu/std
A = mu ** 2 / (2 * (std ** 2))
B = (y_predictor ** 2 / (-2 * (std ** 2))) + (-0.5 * log(2 * pi * (std ** 2)))

# Create a figure and a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# T vs. y
axs[0, 0].plot(y_predictor, T, color='red')
axs[0, 0].set_title('T vs. y')
axs[0, 0].set_xlabel('y')
axs[0, 0].set_ylabel('T')
axs[0, 0].grid(True)

# eta vs. y
axs[0, 1].plot(y_predictor, np.full_like(y_predictor, eta), color='red')
axs[0, 1].set_title('eta vs. y')
axs[0, 1].set_xlabel('y')
axs[0, 1].set_ylabel('eta')
axs[0, 1].grid(True)

# A vs. y
axs[0, 2].plot(y_predictor, np.full_like(y_predictor, A), color='red')
axs[0, 2].set_title('A vs. y')
axs[0, 2].set_xlabel('y')
axs[0, 2].set_ylabel('A')
axs[0, 2].grid(True)

# B vs. y
axs[1, 0].plot(y_predictor, B, color='red')
axs[1, 0].set_title('B vs. y')
axs[1, 0].set_xlabel('y')
axs[1, 0].set_ylabel('B')
axs[1, 0].grid(True)

# f vs. y
axs[1, 1].plot(y_predictor, norm.pdf(y_predictor, mu, std), color='red')
axs[1, 1].set_title('f vs. y')
axs[1, 1].set_xlabel('y')
axs[1, 1].set_ylabel('f')
axs[1, 1].grid(True)

# F vs. y
axs[1, 2].plot(y_predictor, norm.cdf(y_predictor, mu, std), color='red')
axs[1, 2].set_title('F vs. y')
axs[1, 2].set_xlabel('y')
axs[1, 2].set_ylabel('F')
axs[1, 2].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 2

# Initialize variables
n = 10
p = 0.7
y_predictor = np.arange(0, 11, 1)

# Calculations
T = y_predictor
eta = log( 0.84)
A = -n * log(1-p)
B = [log(comb(n, k)) for k in y_predictor]

# Create a figure and a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# T vs. y
axs[0, 0].plot(y_predictor, T, color='red')
axs[0, 0].set_title('T vs. y')
axs[0, 0].set_xlabel('y')
axs[0, 0].set_ylabel('T')
axs[0, 0].grid(True)

# eta vs. y
axs[0, 1].plot(y_predictor, np.full_like(y_predictor, eta), color='red')
axs[0, 1].set_title('eta vs. y')
axs[0, 1].set_xlabel('y')
axs[0, 1].set_ylabel('eta')
axs[0, 1].grid(True)

# A vs. y
axs[0, 2].plot(y_predictor, np.full_like(y_predictor, A), color='red')
axs[0, 2].set_title('A vs. y')
axs[0, 2].set_xlabel('y')
axs[0, 2].set_ylabel('A')
axs[0, 2].grid(True)

# B vs. y
axs[1, 0].plot(y_predictor, B, color='red')
axs[1, 0].set_title('B vs. y')
axs[1, 0].set_xlabel('y')
axs[1, 0].set_ylabel('B')
axs[1, 0].grid(True)

# f vs. y (PDF using stem)
axs[1, 1].stem(y_predictor, binom.pmf(y_predictor, n, p), 'r-', basefmt=" ", use_line_collection=True)
axs[1, 1].set_title('f vs. y')
axs[1, 1].set_xlabel('y')
axs[1, 1].set_ylabel('f')
axs[1, 1].grid(True)

# F vs. y (CDF using stem)
axs[1, 2].stem(y_predictor, binom.cdf(y_predictor, n, p), 'r-', basefmt=" ", use_line_collection=True)
axs[1, 2].set_title('F vs. y')
axs[1, 2].set_xlabel('y')
axs[1, 2].set_ylabel('F')
axs[1, 2].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 3

# Initialize variables
lammbda = 5
y_predictor = np.arange(0, 21, 1)

# Calculations
T = y_predictor
eta = log10(lammbda)
A = lammbda
B = [-log10(factorial(element)) for element in y_predictor]

# Create a figure and a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# T vs. y
axs[0, 0].plot(y_predictor, T, color='red')
axs[0, 0].set_title('T vs. y')
axs[0, 0].set_xlabel('y')
axs[0, 0].set_ylabel('T')
axs[0, 0].grid(True)

# eta vs. y
axs[0, 1].plot(y_predictor, np.full_like(y_predictor, eta), color='red')
axs[0, 1].set_title('eta vs. y')
axs[0, 1].set_xlabel('y')
axs[0, 1].set_ylabel('eta')
axs[0, 1].grid(True)

# A vs. y
axs[0, 2].plot(y_predictor, np.full_like(y_predictor, A), color='red')
axs[0, 2].set_title('A vs. y')
axs[0, 2].set_xlabel('y')
axs[0, 2].set_ylabel('A')
axs[0, 2].grid(True)

# B vs. y
axs[1, 0].plot(y_predictor, B, color='red')
axs[1, 0].set_title('B vs. y')
axs[1, 0].set_xlabel('y')
axs[1, 0].set_ylabel('B')
axs[1, 0].grid(True)

# f vs. y (PDF using stem)
axs[1, 1].stem(y_predictor, poisson.pmf(y_predictor, lammbda), 'r-', basefmt=" ", use_line_collection=True)
axs[1, 1].set_title('f vs. y')
axs[1, 1].set_xlabel('y')
axs[1, 1].set_ylabel('f')
axs[1, 1].grid(True)

# F vs. y (CDF using stem)
axs[1, 2].stem(y_predictor, poisson.cdf(y_predictor, lammbda ), 'r-', basefmt=" ", use_line_collection=True)
axs[1, 2].set_title('F vs. y')
axs[1, 2].set_xlabel('y')
axs[1, 2].set_ylabel('F')
axs[1, 2].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
